# Exercise 1: Linear Regression

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/toche7/DataSets/raw/main/sale.txt', header = None, sep = ',')
```

```python
df.columns = ['pop','sale']
```

```python
df
```

```python
X = df[['pop']]
X.head()
```

```python
y = df[['sale']]
y.tail()
```

```python
from sklearn.linear_model import LinearRegression

regModel = LinearRegression()
```

```python
regModel.fit(X,y)
```

```python
regModel.score(X,y)
```

```python
yhat = regModel.predict(X)
yhat
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,y,'*')
plt.plot(X,yhat,'^')
```
